# 01. マスタデータの作成と確認

業種マスタ・地域マスタの読み込みと分布の確認を行う。

## 目的
- `data/master/industry_master.json` の内容確認
- `data/master/region_master.json` の内容確認
- 仕様書で定義した分布パラメータの可視化

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'sans-serif'

# マスタデータの読み込み
with open('../data/master/industry_master.json', 'r', encoding='utf-8') as f:
    master = json.load(f)

with open('../data/master/region_master.json', 'r', encoding='utf-8') as f:
    regions = json.load(f)

print('業種数:', len(master['industries']))
print('都道府県数:', sum(len(r['prefectures']) for r in regions['regions']))
```

```python
# 業種分布の確認
df_ind = pd.DataFrame(master['industries'])
df_ind[['name', 'target_count', 'ratio']]
```

```python
# 売上規模分布の確認
df_rev = pd.DataFrame(master['revenue_bands'])
df_rev
```

```python
# 譲渡理由分布の確認
df_tr = pd.DataFrame(master['transfer_reasons'])
df_tr
```
